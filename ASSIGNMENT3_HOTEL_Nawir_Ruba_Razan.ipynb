Summerization:

```python
from transformers import pipeline
import pandas as pd
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
df_combined_reviews = df.groupby(['hotel_name', 'locality', 'country', 'rating_value' ])['review_title'].apply(list).reset_index(name='final_rev')
df_combined_reviews
```

```python
summaries = []
for i in df_combined_reviews['final_rev']:
    ARTICLE = ' '.join(i) if isinstance(i, list) else i  # Ensure ARTICLE is a string
    input_length = len(ARTICLE.split())  # Calculate the number of words in the input
    max_new_tokens = max(1, input_length - 1)  # Generate up to input_length - 1 tokens
    summary = summarizer(ARTICLE, max_new_tokens=max_new_tokens, do_sample=False)[0]['summary_text']
    summaries.append(summary)
```

```python
df_combined_reviews['New_summery'] = summaries
```

```python
print(df_combined_reviews.columns)
```

```python
df_combined_reviews['final_rev'] = ' Hotel name: ' + df_combined_reviews['hotel_name'] + " Review: " + df_combined_reviews['New_summery'] + " Locality: " + df_combined_reviews['locality'] + " Country: " + df_combined_reviews['country'] + " Rating value: " + df_combined_reviews['rating_value'].astype(str)
```

Search Engine:

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("nomic-ai/nomic-embed-text-v1", trust_remote_code=True)
sentences = df_combined_reviews['final_rev'].tolist()
embeddings = model.encode(sentences)
```

```python
!pip install faiss-Gpu
import faiss
```

```python
d = embeddings.shape[1]
d
```

```python
nlist = 100
quantizer = faiss.IndexFlatL2(d)
index = faiss.IndexIVFFlat(quantizer, d, nlist)
index.is_trained
```

```python
index.train(embeddings)
index.is_trained
```

```python
index.add(embeddings)
index.ntotal
```

```python
k = 10
xq = model.encode(['luxury hotels in san francisco with kids safe pool'])
```

```python
D, I = index.search(xq, k)
```

```python
from pprint import pprint
pprint({sentences[i] for i in I[0]})
```

```python
def search(query):
  xq = model.encode([query])
  D, I = index.search(xq, k)
  return [f'{sentences[i]}' for i in I[0]]
```

```python
def greet(query):
  k = 10
  xq = model.encode([query])
  result = search(query)
  return result
```
