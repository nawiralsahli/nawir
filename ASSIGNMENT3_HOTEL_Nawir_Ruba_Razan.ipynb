```python
from datasets import load_dataset

ds = load_dataset("traversaal-ai-hackathon/hotel_datasets")
```

Preprocessing:

```python
import pandas as pd

df = pd.DataFrame(ds['train'])
df.to_csv('hotel_datasets.csv', index=False)
```

```python
for i in range(len(df['country'])):
  if df['country'][i] == 'Türkiye':
    df['country'][i] = 'Turkiye'
```

```python
df['country'].unique()
```

```python
df.drop_duplicates(inplace=True)
```

Summerization:

```python
from transformers import pipeline
import pandas as pd
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
df_combined_reviews = df.groupby('hotel_name')['review_title'].apply(lambda reviews: ' '.join(reviews)).reset_index()
df_combined_reviews.rename(columns={'review': 'combined_reviews'}, inplace=True)
df_combined_reviews['locality'] = df['locality']
df_combined_reviews['country'] = df['country']
df_combined_reviews['rating_value'] = df['rating_value']
```

```python
df_combined_reviews
```

```python
summaries = []
for i in df_combined_reviews['review_title']:
  ARTICLE = i
  summary = summarizer(ARTICLE, max_length=100, min_length=30, do_sample=False)[0]['summary_text']
  summaries.append(summary)
```

```python
df_combined_reviews['New_summery'] = summaries
```

```python
df_combined_reviews['final_review'] = ' Hotel name: ' + df_combined_reviews['hotel_name'] + " Review: " + df_combined_reviews['New_summery'] + " Locality: " + df_combined_reviews['locality'] + " Country: " + df_combined_reviews['country'] + " Rating value: " + df_combined_reviews['rating_value'].astype(str)
```

Search Engine:

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("nomic-ai/nomic-embed-text-v1", trust_remote_code=True)
sentences = df_combined_reviews['final_review'].tolist()
embeddings = model.encode(sentences)
```

```python
!pip install faiss-Gpu
import faiss
```

```python
d = embeddings.shape[1]
d
```

```python
nlist = 100
quantizer = faiss.IndexFlatL2(d)
index = faiss.IndexIVFFlat(quantizer, d, nlist)
index.is_trained
```

```python
index.train(embeddings)
index.is_trained
```

```python
index.add(embeddings)
index.ntotal
```

```python
k = 10
xq = model.encode(['luxury hotels in san francisco with kids safe pool'])
```

```python
D, I = index.search(xq, k)
```

```python
from pprint import pprint
pprint([f'{i}: {sentences[i]}' for i in I[0]])
```

```python
def search(query):
  xq = model.encode([query])
  D, I = index.search(xq, k)
  return [f'{sentences[i]}' for i in I[0]]
```

```python
def greet(query):
  k = 10
  xq = model.encode([query])
  result = search(query)
  return result
```
